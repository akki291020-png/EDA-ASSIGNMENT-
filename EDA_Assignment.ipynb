Question 2: Check for missing values in all columns and describe your approach for handling them.


```python
import pandas as pd

# Load the DataFrame
df = pd.read_csv('BIKE DETAILS.csv')

# Calculate the number of missing values per column
missing_values_count = df.isnull().sum()

# Calculate the total number of rows
total_rows = len(df)

# Calculate the percentage of missing values per column
missing_values_percent = (missing_values_count / total_rows) * 100

# Create a summary DataFrame
missing_summary = pd.DataFrame({
    'Missing Count': missing_values_count,
    'Missing Percentage (%)': missing_values_percent
})

# Display the summary, showing only columns with missing values
print("Missing Value Summary (Columns with Missing Data):")
print(missing_summary[missing_summary['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False))

# Display the full summary
print("\n" + "="*50 + "\n")
print("Missing Value Summary (All Columns):")
print(missing_summary.sort_values(by='Missing Count', ascending=False))
```

Question 4: Create a bar plot to visualize the average selling price for each seller_type and write one observation.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame
df = pd.read_csv('BIKE DETAILS.csv')

# Calculate the average selling price for each seller type
average_price_by_seller = df.groupby('seller_type')['selling_price'].mean().sort_values(ascending=False).reset_index()

# Set the aesthetic style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(8, 5))
sns.barplot(
    x='seller_type',
    y='selling_price',
    data=average_price_by_seller,
    palette='viridis'
)

# Add labels and title
plt.title('Average Selling Price by Seller Type', fontsize=14)
plt.xlabel('Seller Type', fontsize=12)
plt.ylabel('Average Selling Price (in Rupees)', fontsize=12)
plt.ticklabel_format(style='plain', axis='y') # Prevent scientific notation on y-axis

# Add text labels on top of the bars for exact values
for index, row in average_price_by_seller.iterrows():
    plt.text(
        index,
        row['selling_price'] + 1000,
        f'₹{row["selling_price"]:,.0f}',
        color='black',
        ha="center",
        fontsize=10
    )

# Show the plot (the system will generate the plot preview)
plt.show()

# Print the calculated average prices
print("Average Selling Price by Seller Type:")
print(average_price_by_seller)
```

Question 6: Use the IQR method to detect and remove outliers from the km_driven column. Show before-and-after summary statistics.


```python
import pandas as pd

# Load the DataFrame
df = pd.read_csv('BIKE DETAILS.csv')

# --- Summary Statistics BEFORE Outlier Removal ---
print("--- Summary Statistics for km_driven (BEFORE Outlier Removal) ---")
km_driven_before = df['km_driven'].describe().to_frame()
print(km_driven_before)
print("\n" + "="*70 + "\n")

# --- IQR Method Implementation ---
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier bounds (k=1.5)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create the filtered DataFrame (removing outliers)
# Note: Since the lower bound is negative, we only filter based on the upper bound.
df_clean = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)].copy()

# --- Summary Statistics AFTER Outlier Removal ---
print("--- Summary Statistics for km_driven (AFTER Outlier Removal) ---")
km_driven_after = df_clean['km_driven'].describe().to_frame()
print(km_driven_after)
print("\n" + "="*70 + "\n")

# --- Report Outlier Counts and Bounds ---
outliers_count = len(df) - len(df_clean)
print(f"Calculated Q1: {Q1:,.0f}")
print(f"Calculated Q3: {Q3:,.0f}")
print(f"Calculated IQR: {IQR:,.0f}")
print(f"Upper Bound (Q3 + 1.5*IQR): {upper_bound:,.0f}")
print(f"Number of outliers removed (rows dropped): {outliers_count}")
```

Question 7: Create a scatter plot of year vs. selling_price to explore the relationship between a bike's age and its price.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame
df = pd.read_csv('BIKE DETAILS.csv')

# Set the aesthetic style
sns.set_style("whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='year',
    y='selling_price',
    data=df,
    alpha=0.6,
    color='darkblue'
)

# Add labels and title
plt.title('Relationship between Bike Manufacturing Year and Selling Price', fontsize=14)
plt.xlabel('Manufacturing Year', fontsize=12)
plt.ylabel('Selling Price (in Rupees)', fontsize=12)
plt.ticklabel_format(style='plain', axis='y') # Prevent scientific notation on y-axis

# Ensure x-axis ticks are integer years
plt.xticks(sorted(df['year'].unique()), rotation=45, ha='right', fontsize=10)

# Save the plot
plt.savefig('year_vs_selling_price_scatter_plot.png')
plt.close()
```

OutputObservationThe scatter plot clearly shows a positive correlation between the manufacturing year and the selling price:Newer bikes (higher year values) are concentrated at the higher selling prices.Older bikes (lower year values) are predominantly found at the lower selling prices.The prices of the most expensive bikes (those above $\text{Rs. }400,000$) are exclusively found among the newest models (around $\text{2017}$ to $\text{2020}$).The variance in selling prices is much wider for newer bikes than for older bikes, suggesting a greater variety in bike segments (e.g., commuter vs. premium) among recent models.

Question 8: Convert the seller_type column into numeric format using one-hot encoding. Display the first 5 rows of the resulting DataFrame.


```python
import pandas as pd

# Load the DataFrame
df = pd.read_csv('BIKE DETAILS.csv')

# --- 1. Perform One-Hot Encoding on 'seller_type' ---
# Create new binary columns for each category in 'seller_type'
seller_dummies = pd.get_dummies(df['seller_type'], prefix='seller', drop_first=False)

# --- 2. Concatenate the new columns and drop the original 'seller_type' column ---
df_encoded = pd.concat([df.drop('seller_type', axis=1), seller_dummies], axis=1)

# --- 3. Display the first 5 rows of the resulting DataFrame ---
print("First 5 rows of the DataFrame after One-Hot Encoding 'seller_type':")
print(df_encoded.head(5))

# --- 4. Display the columns and their data types to confirm the change ---
print("\n" + "="*70 + "\n")
print("DataFrame info after encoding:")
print(df_encoded.info())
```

Question 9: Generate a heatmap of the correlation matrix for all numeric columns.What correlations stand out the most?

This report summarizes the exploratory data analysis (EDA) performed on the Bike Details dataset, identifying the most significant factors affecting a bike's selling price and detailing the data preparation steps taken.

Key Factors Affecting a Bike's Selling Price
Based on correlation analysis and visualization, the three most important factors affecting a bike's selling price are:

Ex-Showroom Price (Original Cost): This is the most critical factor, showing an extremely strong positive correlation (r≈0.92) with the selling price. The original value of the bike dictates its baseline used price.

Year of Manufacture (Age): The bike's age, represented by the year column, shows a moderate positive correlation (r≈0.40). Newer bikes (higher year) generally command higher prices, which is the direct result of depreciation.

Kilometers Driven (Mileage): The km_driven shows a weak negative correlation (r≈−0.21). While not as strong as the price or age, higher mileage contributes to a lower selling price due to increased wear and tear.

Other Relevant Findings:
Seller Type: Bikes sold by Individual sellers have a higher average selling price (Rs. 59,712) compared to those sold by Dealers (Rs. 46,667). This suggests that individuals might be listing more high-value or premium bikes, or dealers might offer more competitive pricing.

Price Distribution: The selling price distribution is highly right-skewed, with the vast majority of sales clustered below Rs. 70,000, and a long tail representing a small number of expensive, premium bikes.

Data Cleaning and Feature Engineering Performed
Outlier Handling (Data Cleaning):

Action: Outliers were detected and removed from the km_driven column using the Interquartile Range (IQR) method (1.5×IQR).

Impact: This process removed 39 rows and significantly reduced the mean (≈34,360 km to ≈28,203 km) and standard deviation, making the data more robust and representative of typical bike usage.

Action: Checked for missing values.

Action: The categorical feature seller_type was converted into a numeric format using One-Hot Encoding (creating binary columns: seller_Dealer and seller_Individual).
